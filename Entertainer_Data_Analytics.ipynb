```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("/content/Entertainer - Breakthrough Info.xlsx")
```

```python
mean=df["Year of First Oscar/Grammy/Emmy"].mean()
# Fill missing values with a placeholder (e.g., 0 or a specific year)
df['Year of First Oscar/Grammy/Emmy'].fillna(mean, inplace=True)
```

```python
print(df.isnull().sum())
```

# Year-wise trends for breakthroughs and awards

```python
plt.figure(figsize=(14, 7))
sns.countplot(data=df, x='Year of Breakthrough/#1 Hit/Award Nomination')
plt.title('Number of Breakthroughs and Major Awards by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# Scatter plot of breakthrough year vs. award year

```python
plt.figure(figsize=(14, 7))
sns.scatterplot(data=df, x='Year of Breakthrough/#1 Hit/Award Nomination', y='Year of First Oscar/Grammy/Emmy')
plt.title('Breakthrough Year vs. Award Year')
plt.xlabel('Year of Breakthrough')
plt.ylabel('Year of First Major Award')
plt.show()
```

Trend Analysis

```python
# Year-wise trends for breakthroughs and awards
plt.figure(figsize=(14, 7))

# Plot for Breakthroughs
sns.countplot(data=df, x='Year of Breakthrough/#1 Hit/Award Nomination', palette='viridis')
plt.title('Number of Breakthroughs by Year')
plt.xlabel('Year of Breakthrough')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Plot for Major Awards
plt.figure(figsize=(14, 7))
sns.countplot(data=df, x='Year of First Oscar/Grammy/Emmy', palette='viridis')
plt.title('Number of Major Awards by Year')
plt.xlabel('Year of First Major Award')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

 Career Longevity Analysis


```python
# Calculate time to major award
df['Time to Award'] = df['Year of First Oscar/Grammy/Emmy'] - df['Year of Breakthrough/#1 Hit/Award Nomination']

# Distribution of time to award
plt.figure(figsize=(14, 7))
sns.histplot(df['Time to Award'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Time to Major Award')
plt.xlabel('Years to Major Award')
plt.ylabel('Frequency')
plt.show()

# Boxplot to show spread and outliers
plt.figure(figsize=(14, 7))
sns.boxplot(data=df, x='Time to Award', color='lightgreen')
plt.title('Boxplot of Time to Major Award')
plt.xlabel('Years to Major Award')
plt.show()
```

Correlation Analysis

```python
# Calculate correlation
correlation_matrix = df[['Year of Breakthrough/#1 Hit/Award Nomination', 'Year of First Oscar/Grammy/Emmy']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0, fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

 Top Entertainers Analysis

```python
# Top entertainers by number of breakthroughs
top_breakthroughs = df['Entertainer'].value_counts().head(10)
print("Top 10 Entertainers by Number of Breakthroughs:")
print(top_breakthroughs)

# Plot for top entertainers
plt.figure(figsize=(14, 7))
sns.barplot(x=top_breakthroughs.index, y=top_breakthroughs.values, palette='viridis')
plt.title('Top 10 Entertainers by Number of Breakthroughs')
plt.xlabel('Entertainer')
plt.ylabel('Number of Breakthroughs')
plt.xticks(rotation=45)
plt.show()
```
